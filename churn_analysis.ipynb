```python
import pandas as pd
```

```python
df = pd.read_csv('analyses_customer_churn_features.csv')
```

```python
df.info()  # Display information about the DataFrame
```

```python
df.head()  # Display the first few rows of the DataFrame
```

```python
df.describe()  # Get a statistical summary of the DataFrame
```

```python
# Análise de correlação entre as colunas numéricas e 'is_churn'
correlations = df.corr(numeric_only=True)['is_churn'].drop('is_churn').sort_values(ascending=False)

print("Correlação das variáveis com 'is_churn':")
print(correlations)

# Gráfico de barras das correlações
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
correlations.plot(kind='bar')
plt.title("Correlação das variáveis numéricas com 'is_churn'")
plt.ylabel("Coeficiente de Correlação")
plt.xlabel("Variável")
plt.tight_layout()
plt.show()
```
